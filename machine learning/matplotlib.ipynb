### Object-oriented interface

````python
# First create a grid of plots
# ax will be an array of two Axes objects
# 多图绘制，一张画布绘制多个子图
fig, ax = plt.subplots(2) # 创建两个canvas对象，创建两个子图
# Call plot() method on the appropriate object 
# 获取每个子图对象
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

````python
#多图绘制
#导入包numpy
import numpy as np
#导入matplotlib的pyplot模块
import matplotlib.pyplot as plt
#定义一维数组
old_version = np.arange(5)
new_version = np.arange(5)
line1 = range(0, len(old_version))
line2 = range(0, len(new_version))
diff = old_version - new_version
line3 = range(0, len(diff))

fig, ax = plt.subplots() # 创建图实例
ax.plot(line1, old_version, label='old_version') # 作y1 = x 图，并标记此线名为linear  # 给线添加标注
ax.plot(line2, new_version, label='new_version') #作y2 = x^2 图，并标记此线名为quadratic
ax.plot(line3, diff, label='diff') # 作y3 = x^3 图，并标记此线名为cubic

ax.set_xlabel('x label') #设置x轴名称 x label
ax.set_ylabel('y label') #设置y轴名称 y label
ax.set_title('Simple Plot') #设置图名为Simple Plot
ax.legend() #自动检测要在图例中显示的元素，并且显示
plt.show() #图形可视化
````

````python
'''
数据处理，统计学处理，曲线滤波，均值滤波，线性滤波
https://blog.csdn.net/weixin_42782150/article/details/107176500 
https://blog.csdn.net/sinat_28252525/article/details/80462437 
为什么不直接使用excel呢？点点点就完了

'''
import numpy as np
import scipy
y = np.random.random(size=100)
# 使用Savitzky-Golay 滤波器后得到平滑图线
from scipy.signal import savgol_filter
y_smooth = scipy.signal.savgol_filter(y,53,3)  
# 亦或
y_smooth2 = savgol_filter(y, 99, 1, mode= 'nearest')
# 备注：
# y：代表曲线点坐标（x,y）中的y值数组
# window_length：窗口长度，该值需为正奇整数。例如：此处取值53
# k值：polyorder为对窗口内的数据点进行k阶多项式拟合，k的值需要小于window_length。例如：此处取值3
# mode：确定了要应用滤波器的填充信号的扩展类型。（This determines the type of extension to use for the padded signal to which the filter is applied. ）

import matplotlib.pyplot as plt
plt.plot(np.arange(len(y)), y_smooth, '*', np.arange(len(y)), y_smooth2, '-') # 指定线条类型，多图绘制，将拖个线条绘制到同一个图中
plt.show()
````

### Bar plots

````python
from scipy.stats import norm
# 多图绘制，绘制在一张图上。在一幅图上绘制多个柱状图
X1 = np.arange(0,20,1)
Y1 = norm(5,2).pdf(X1) # 正态分布随机采样
plt.bar(X1,Y1,facecolor='b'); # 设置柱状图的颜色

X2 = np.arange(-10,10,1)
Y2 = norm(0,5).pdf(X2)
plt.bar(X2,Y2,facecolor='r', alpha=.8); # 设置柱状图的透明度
````

### Defining functions and plotting them

````python
def circle_area(r):
    return np.pi*r*r

def sphere_volume(r):
    return (4/3)*np.pi*r*r*r

r = np.arange(0,5,0.1)
# 绘制公式
plt.plot(r, circle_area(r), label=r'$A=\pi r^2$');
plt.plot(r, sphere_volume(r), label=r'$V=\frac{4}{3} \pi r^3$'); # 添加公式到图表中

plt.legend(loc='upper center')
````

### Writing text in plots

````python
plt.plot(r, circle_area(r), label='circle surface');
plt.plot(r, sphere_volume(r), label='sphere volume');

plt.xlabel("r (radius)")
plt.ylabel("V (volume) / A (area)")
plt.title('The relationship between circle surface area and sphere volume')

xpos = 4
plt.text(xpos, circle_area(xpos) + 30, r'$A=\pi r^2$')
# python会将最后一个参数解释为一种公式表示语言，LaTex http://www.uinio.com/Math/LaTex/
plt.text(xpos, sphere_volume(xpos) + 100, r'$V=\frac{4}{3} \pi r^3$') 
````

````python
# 找到交差点
def ps_equal (p1,p2):
    return np.abs(p1-p2) < 1e-5
def f1(t):
    return (2*t) + 1
def f2(t):
    return (-2*t) + 1
x = np.arange (-10,10,0.01)
plt.plot (x, f1(x), 'r')
plt.plot (x, f2(x), 'b')
for el in x:
    if ps_equal(f1(el), f2(el)):
        plt.text(el, f1(el), r"This is the position!!")
````

### Instead of adding a text, we can annotate it!

````python
def ps_equal (p1,p2):
    return np.abs(p1-p2) < 1e-5

def f1(t):
    return (2*t) + 1

def f2(t):
    return (-2*t) + 1

x = np.arange (-10,10,0.01)

plt.plot (x, f1(x), 'r')
plt.plot (x, f2(x), 'b')

for el in x:
    if ps_equal(f1(el), f2(el)): # 添加箭头指示，添加注释
        plt.annotate('Intersection point', xy=(0, 0), xytext=(-5, -10),
            arrowprops=dict(facecolor='black')
            )
````

### A better way for creating aligned subplots

The function `plt.subplots()` takes several arguments, but the most relevant ones are `nrows` and `ncols` (number of rows and number of columns).

This function returns a plt.Figure object and a list of axes (one element for each subplot).

````python
nrows = 2
ncols = 3
for i in range(1, nrows*ncols+1):
    plt.subplot(nrows, ncols, i) # 创建并且激活某一个子图，设置字体大小
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
````

````python
plt.subplots_adjust(hspace=0.0,wspace=0.1) # 设置水平和数值间距
nrows = 2
ncols = 3
for i in range(1, nrows*ncols+1):
    plt.subplot(nrows, ncols, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize=18, ha='center')
````

````python
'''
==============
3D scatterplot
==============

Demonstration of a basic scatterplot in 3D.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

ax.text(26,40,-5, "Hi matplotlib3d");
````
